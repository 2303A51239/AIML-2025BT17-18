```python

import re
from collections import Counter

file_path = "/content/dataset.txt"
with open(file_path, "r", encoding="utf-8") as file:
    data = file.read()

complaints_keywords = [
    "kapanma", "yavaş", "sorun", "hata", "fan sesi",
    "garanti", "paketleme", "şarj", "güncelleme", "klavye"
]

complaints = re.findall(rf"\b({'|'.join(complaints_keywords)})\b", data, re.IGNORECASE)
complaints_count = Counter(complaints)

top_complaints = complaints_count.most_common(5)
print("Top 5 Complaints and their frequencies:")
for issue, freq in top_complaints:
    print(f"{issue}: {freq}")
```

```python
import re
from collections import defaultdict

file_path = "/content/dataset.txt"
with open(file_path, "r", encoding="utf-8") as file:
    data = file.readlines()

ratings = defaultdict(list)
for line in data:
    match = re.search(r"\b(\d(\.\d)?) yıldız\b", line, re.IGNORECASE)
    if match:
        rating = float(match.group(1))
        ratings[rating].append(line.strip())

similar_rating_counts = {rating: len(products) for rating, products in ratings.items()}
products_above_4 = [(rating, products) for rating, products in ratings.items() if rating > 4]

print("Product Counts by Rating:")
for rating, count in sorted(similar_rating_counts.items()):
    print(f"Rating {rating}: {count} products")

print("\nProducts with >4 Ratings:")
for rating, products in products_above_4:
    print(f"\nRating {rating}:")
    for product in products:
        print(product)
```

```python

import re

file_path = "/content/dataset.txt"
with open(file_path, "r", encoding="utf-8") as file:
    data = file.readlines()

ratings = []
for line in data:
    match = re.search(r"\b(\d(\.\d)?) yıldız\b", line, re.IGNORECASE)
    if match:
        rating = float(match.group(1))
        ratings.append(rating)

if ratings:
    max_rating = max(ratings)
    min_rating = min(ratings)
    print(f"Maximum Rating: {max_rating}")
    print(f"Minimum Rating: {min_rating}")
else:
    print("No ratings found.")
```

```python

import re
from collections import Counter

file_path = "/content/dataset.txt"
with open(file_path, "r", encoding="utf-8") as file:
    data = file.readlines()

product_mentions = []
for line in data:
    match = re.search(r"(?i)([a-zA-Z0-9]+)\s*[^\n]*", line)
    if match:
        product_mentions.append(match.group(1))

most_reviewed_product = Counter(product_mentions).most_common(1)

if most_reviewed_product:
    product, count = most_reviewed_product[0]
    print(f"Most Reviewed Product: {product} ({count} reviews)")
else:
    print("No products found.")
```

```python
import re

file_path = "/content/dataset.txt"
with open(file_path, "r", encoding="utf-8") as file:
    data = file.readlines()

electronics_keywords = ["laptop", "phone", "tablet", "computer", "tv", "headphone", "camera", "printer"]
edible_keywords = ["food", "snack", "chocolate", "drink", "coffee", "tea", "fruit", "vegetable", "candy"]

electronics_reviews = []
edible_reviews = []

for line in data:
    if any(keyword.lower() in line.lower() for keyword in electronics_keywords):
        electronics_reviews.append(line.strip())
    elif any(keyword.lower() in line.lower() for keyword in edible_keywords):
        edible_reviews.append(line.strip())

print("Reviews for Electronic Products:")
for review in electronics_reviews:
    print(review)

print("\nReviews for Edible Products:")
for review in edible_reviews:
    print(review)
```
